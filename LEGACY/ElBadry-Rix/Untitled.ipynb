```python
def experiment(sample,k=0.37,printit=False):
    N = sample.size
    best=0
    i = 0
    found_best = False
    while (i < k*N):
        best = np.maximum(best,sample[i])
        print(i,best,sample[i]) if printit else 0
        i+=1
    while(i < N-1):
        if (best < sample[i]):
            print('Found best: ',i,best,sample[i]) if printit else 0
            break
        i+=1
    print('Last value:',sample[i]) if printit else 0
    return sample[i]
```

```python
results = np.array([])
k=0.05
#experiment(np.random.randint(1,100,100),printit=True,k=0.05)
for i in range(999):
    results = np.append(results,experiment(np.random.rand(1000),k))
plt.hist(results,bins=100)
```
