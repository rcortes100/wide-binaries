```python
je = pd.read_pickle('../data_processed/jim_es_final_selection.pkl')

je['dist_mod_x'] = -5*(np.log10(1E-3*je.parallax_x)+1)
je['g_abs_x']    = je.phot_g_mean_mag_x - je.dist_mod_x

je['dist_mod_y'] = -5*(np.log10(1E-3*je.parallax_y)+1)
je['g_abs_y']   = je.phot_g_mean_mag_y - je.dist_mod_y
```

```python
fig, ax2 = plt.subplots()
fnt = 10
dotsize = 1
ax2.invert_yaxis()

ax2.set_xlabel('G$_{BP}$-G$_{RP}$',fontsize=fnt)
#ax2.set_ylabel('M$_G$',fontsize=fnt,labelpad=0)

#ax2.plot(subtab.BP_RP1,subtab.G_abs1, 
ax2.plot(je.bp_rp_x,je.g_abs_x,
         'o',ms=dotsize, label='$M_G$ pri')
#ax2.plot(subtab.BP_RP2,subtab.G_abs2 +3,
ax2.plot(je.bp_rp_y,je.g_abs_y +3, 
         'o',ms=dotsize, label='$M_G$ sec +3')
ax2.legend(markerscale=1)

plt.show()
```
