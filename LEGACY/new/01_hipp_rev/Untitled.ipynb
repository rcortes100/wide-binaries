```python
def matA(ra,dec):
    return np.array([[-np.sin(ra)            , np.cos(ra)             , 0.],
                     [-np.cos(ra)*np.sin(dec), -np.sin(ra)*np.sin(dec), np.cos(dec)],
                     [ np.cos(ra)*np.cos(dec),  np.sin(ra)*np.cos(dec), np.sin(dec)]])
def matB(ra,dec):
    return np.array([[-np.sin(ra)            , -np.cos(ra)*np.sin(dec), np.cos(ra)*np.cos(dec)],
                     [ np.cos(ra)            , -np.sin(ra)*np.sin(dec), np.sin(ra)*np.cos(dec)],
                     [                     0.,             np.cos(dec), np.sin(dec)]])

k   = 4.74
#k   =  48.101

rap   = (np.pi/180)*114.9970833 #radians
decp  = (np.pi/180)*(-3.5975) #radians
pp    =  70.44 #mas
vrp   = -21.8 #km/s
mrap  =  71.7
mdecp = -276.1

Vp = np.array([k*mrap/pp,
               k*mdecp/pp,
               vrp])

ras   = (np.pi/180)*114.0291667
decs  = (np.pi/180)*(-3.11077777)
mras  = 36.3
mdecs = -253.5
ps    = 69.85

Vs = matA(ras,decs) @ matB(rap,decp) @ Vp

(mrap,mdecp,vrp) - Vs
```
